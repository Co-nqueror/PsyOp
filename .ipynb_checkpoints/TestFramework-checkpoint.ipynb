```python
import sqlite3
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt

conn = sqlite3.connect("Dataset.db")

tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table';", conn)
table_names = tables['name'].tolist()
table_names.pop(0) #Removes sqlite_sequence from the list

table_dropdown = widgets.Dropdown(
    options=table_names,
    description='Table 1:',
    value=table_names[0]
)
table_dropdown2 = widgets.Dropdown(
    options=table_names,
    description='Table 2:',
    value=table_names[0]
)

x_axis = widgets.Dropdown(description='Table 1: X-axis:')
y_axis = widgets.Dropdown(description='Table 2: Y-axis:')

plot_button = widgets.Button(description="Generate Plot", button_style="success")

out = widgets.Output()

def update_columns(change):
    df1 = pd.read_sql(f"SELECT * FROM {table_dropdown.value};", conn)
    x_axis.options = df1.columns.tolist()
    
    df2 = pd.read_sql(f"SELECT * FROM {table_dropdown2.value};", conn)
    y_axis.options = df2.columns.tolist()

def generate_plot(b):
    with out:
        clear_output()
        df1 = pd.read_sql(f"SELECT * FROM {table_dropdown.value};", conn)
        df2 = pd.read_sql(f"SELECT * FROM {table_dropdown2.value};", conn)
        if x_axis.value and y_axis.value:
            plt.figure(figsize=(6,4))
            df1.plot(kind='bar', x=x_axis.value, y=y_axis.value, legend=False)
            plt.title(f"{y_axis.value} vs {x_axis.value}")
            plt.show()

table_dropdown.observe(update_columns, names='value')
table_dropdown2.observe(update_columns, names='value')
plot_button.on_click(generate_plot)

update_columns(None)

display(widgets.VBox([table_dropdown, table_dropdown2, x_axis, y_axis, plot_button, out]))
```
